# Critical Region 

# Example 8.1-2 

<p>
We are interested in finding the probability that $\bar{\mathrm{X}} \geq 62.75$ given $\bar{\mathrm{X}} \sim N(60, \frac{10}{\sqrt{52}})$. The result reveals the value of $\alpha$.
</p>

```julia
d = Normal(60, 10 / sqrt(52))
```

```julia
α = ccdf(d, 62.75)
```

<p>We can of course, switch directions</p>

```julia
quantile(d, 1 - α)
```

```julia
c = Normal(62.75, 10 / sqrt(52))
```

```julia
plot(x -> pdf(d, x), 55, 65)
```

# Exercise 8.1-4 

```julia
gum = [7.65, 7.6, 7.65, 7.7, 7.55, 7.55, 7.4, 7.4, 7.5, 7.5];
```

```julia
t = ((mean(gum) - 7.5) / (std(gum) / (sqrt(length(gum)))))
```

```julia
2 * ccdf(TDist(length(gum) - 1), t)
```

```julia
OneSampleTTest(gum, 7.5)
```

# Exercise 8.1-6 

```julia
fvc = [3.4, 3.6, 3.8, 3.3, 3.4, 3.5, 3.7, 3.6, 3.7];
```

```julia
t = (mean(fvc) - 3.4) / (std(fvc) / sqrt(length(fvc)))
```

<p>p-value for a 1-sided test. Note this is below the 0.05 level necessary for 95% confidence</p>

```julia
ccdf(TDist(length(fvc) - 1), t)
```

```julia
OneSampleTTest(fvc, 3.4)
```

```julia
plot(x -> pdf(TDist(length(fvc) - 1), x), -3, 3)
```
