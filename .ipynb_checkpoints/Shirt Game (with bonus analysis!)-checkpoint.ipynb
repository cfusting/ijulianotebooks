# Winning a Special Shirt

![shitboard](http://i.imgur.com/Xc3FSP1.png)

<p>
With the carnival in town exiles are excited at the prospect of winning a special shirt. This year we have a redish style already very popular with Coldy, Melben, and Gawyn!
</p>

<p>
Games of chance are fun, but sometimes it's nice to have an idea of your odds. Given the 75c per play ticket price, even the most experienced c whores (aside from Kalian) will quickly lighten their purses.
</p>

## Rules

<p>
To win the special shirt, the game duck must land on the center square of a 12 x 12 checkered board. It is possible to win other shirts, but we are not concerned with these prospects in this analysis. Assuming the game is fair, exiles have a $\frac{1}{144}$ chance of landing the duck (which seems to move randomly across the board) on the center square.
</p>

![Board](http://i.imgur.com/0AyEyaQ.png)

## Analysis

<p>
Given our odds, it shouldn't be surprising that on average an exile must play 144 times (we will show this later) to win the special shirt. However it is possible to win on your first try, or your 1000th, so what are the odds these events might occur? Let's first consider lady luck who walks up and wins on her first try (has anyone done this?). The probability of this event occuring is exactly the probability of winning the game:
</p> 

```julia
1/144
```

![duck](http://i.imgur.com/VXpJSu0.png)

<p>
A little shy of 1%. Very lucky indeed! Let's consider your average carnival goer who walks up to the game. It's reasonable to expect they have at least 1000c - 2000c in their purse. If they've played some games already, they have some portion remaining and are perhaps are not willing to spend it all on the duck game:
<p>

```julia
1500 / 75
```

<p>
We'll given them 20 tries over the course of their visit. What is the probability they will win a shirt? Consider the probability they loose a game, $\frac{143}{144}$. If they loose 20 times in a row we have $(\frac{143}{144})^{20}$. But we want to know if they win <b>at least</b> once so we take the compliment of the set by subtracting the odds of loosing every time from 1 (the total probability):
</p>

```julia
1 - (143/144)^20
```

<p>
Yikes! Still only ~13% chance to win a shirt for the average carnival going, and this at a cost of 1500c! Note however this is the probability of winning <b> at least </b> one shirt so it is possible the exile will win more. 
</p>
<p>
Let's frame this problem a little more formally so we have access to more powerful math. We can think of this game as a Binomial distribution which will give us the odds of $x$ successes given $n$ trials with a $\frac{1}{144}$ chance to win. Let's first consider our 20 tries example:
</p>

```julia
using Distributions
duckgame20 = Binomial(20, 1/144)
pdf(duckgame20, 1)
```

<p>
But wait, that's slightly lower? That's because the Binomial distribution is giving us the odds of winning <b> exacly </b> once. To calculate the odds of winning at least once we need to add the odds of winning twice, three times, etc. For verbosity I will output the vector of odds in addition to the final value:
</p>

```julia
pdf(duckgame20, 1:20)
```

```julia
sum(pdf(duckgame20, 1:20))
```

<p>
Right, that matches our previous result. Now, how many shirts can a player expect to win in 20 tries?
</p>

```julia
mean(duckgame20)
```

<p>
Not alot of shirt! Earlier we assumed that in 144 tries a player would win one shirt on average. Let's verify this with a Binomial distribution and 144 trials:
</p>

```julia
duckgame144 = Binomial(144, 1/144)
mean(duckgame144)
```

![shirt](http://i.imgur.com/sY0Hv49.png)

<p>
As we expected, one shirt for every 144 tries. That, however, does <b>not</b> imply the probability of winning a shirt in 144 tries is 100%, rather:
</p>

```julia
sum(pdf(duckgame144, 1:144))
```

<p>
Yikes! Those odds are not nearly as nice as we might like for our price tag:
</p>

```julia
144 * 75
```

<p>
Let's plot the probability of winning <b> at least </b> one shirt compared with the number of tries:
</p>

```julia
using Gadfly
set_default_plot_size(24cm, 16cm)
plot(x -> sum(pdf(Binomial(convert(Int, floor(x)), 1/144), 1:x)), 1, 500, 
Guide.xlabel("Tries"),
Guide.ylabel("Probability"),
Guide.title("Probability of Winning a Special Shirt"))
```

<p>
Yikes! It looks like we don't even cross 50% until 100 trials (or 7500c). The more trials the more expensive raising our odds get! So although we can expect to win a shirt every 144 tries on average, this is far from a guarantee.
</p>

```julia
1.4375 / 2
```

<p>
