```julia
using Gadfly, DataFrames
```

```julia
f(β::Float64, s::Float64) = log(1 / (1-s)) * 1 / β * s
```

```julia
plot(
layer(x = [f(1.0, x) for x = .01:.01:1], y = [x for x = .01:.01:1], Geom.line, Theme(default_color=color("orange"))),
layer(x = [f(0.8, x) for x = .01:.01:1], y = [x for x = .01:.01:1], Geom.line, Theme(default_color=color("purple"))),
layer(x = [f(0.5, x) for x = .01:.01:1], y = [x for x = .01:.01:1], Geom.line),
Guide.xlabel("Average k"), Guide.ylabel("S"), Guide.title("Size of Giant Component")
)
```
