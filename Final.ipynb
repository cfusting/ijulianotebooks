```julia
function EstimateProportionSampleSize(ϵ; p = .5, α = .95)
    β = (1 - α) / 2
    z = quantile(Normal(0,1), 1 - β)
    (z^2 * p * (1 - p)) / ϵ^2
end
```

```julia
EstimateProportionSampleSize(.02)
```
